```python
import pandas as pd

# Read in the CSV file
df = pd.read_csv("Totalbak_2023-03-25.csv")

# Group the data by every 8 rows
grouped_data = df.groupby(df.index // 8)

# Aggregate the values in each group into a single row
new_df = grouped_data.agg({"item": list, "value": list})

# Concatenate the lists in the "item_grouped" and "value_grouped" columns into a single row
new_df["item_grouped"] = new_df["item_grouped"].apply(
    lambda x: [elem for sublist in x for elem in sublist])
new_df["value_grouped"] = new_df["value_grouped"].apply(
    lambda x: [elem for sublist in x for elem in sublist])

# Create a new dataframe with the concatenated columns
final_df = pd.DataFrame(
    {"item": new_df["item_grouped"], "value": new_df["value_grouped"]})
final_df
```
