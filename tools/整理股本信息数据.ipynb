# 整理爬虫biding项目total爬取下来的总股本数据

```python
import pandas as pd

filename = "../data/Total2_2023-03-26.csv"

# Read in the CSV file
df = pd.read_csv(filename)

# Group the data by every 8 rows
grouped_data = df.groupby(df.index // 8)
new_df = grouped_data.agg({"item": list, "value": list})
new_df
```

```python
data = dict(zip(new_df["item"][0], new_df["value"][0]))
data
```

```python
df = pd.DataFrame(columns=new_df["item"][0])
# data = dict(zip(new_df["item"][0], new_df["value"][0]))
for index,row in new_df.iterrows():
#     print(row["item"])
    data = dict(zip(row["item"], row["value"]))
    df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)
#     final_df.append(data, ignore_index=True)
# df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)
# # final_df.append(data, ignore_index=True)
# data1 = dict(zip(new_df["item"][1], new_df["value"][1]))
# df = pd.concat([df, pd.DataFrame([data1])], ignore_index=True)


df.info()
```
