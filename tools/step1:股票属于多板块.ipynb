# 由于一个股票属于多个板块，所以需要这部操作

```python
import pandas as pd
from datetime import datetime
```

```python
code_df = pd.read_csv("../data/板块名称_股票对应.csv", index_col=0, dtype={"股票代码":object})
code_df.head()
```

```python
code_df.info()
```

```python
code_df['股票代码'].nunique()
```

```python
code_df['板块名称'].nunique()
```

```python
df = pd.read_csv(
    f"../data/Hist_{datetime.now().strftime('%Y-%m-%d')}.csv", parse_dates=['日期'], index_col=0, dtype={"股票代码": object})
df.head()
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.drop(columns=['板块名称'],inplace=True)
```

## 获得全部代码的列表

```python
merged_df = pd.merge(df, code_df, on='股票代码')
merged_df
```

```python
merged_df.info()
```

```python
merged_df.to_csv(f"../data/merge_{datetime.now().strftime('%Y-%m-%d')}.csv",index=False)
```
