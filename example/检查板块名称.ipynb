```python

merge_df = pd.read_csv(
    "../data/merge_2023-03-25.csv", parse_dates=['日期'], index_col=0, dtype={"股票代码": object})
value = pd.read_csv("../data/总股本.csv", index_col=0, dtype={"股票代码": object})
value_dict = value['总股本'].to_dict()
merge_df['总股本'] = merge_df['股票代码'].apply(lambda x: value_dict.get(x))
merge_df['总市值'] = merge_df['总股本']*merge_df['收盘']
merge_df.fillna(0, inplace=True)
merge_df.info()
```
